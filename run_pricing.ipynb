```python
import numpy as np
import matplotlib.pyplot as plt
from pricing.black_scholes import black_scholes_price

S = 100
T = 1
r = 0.05
sigma = 0.2
option_type = 'call'
K_values =  np.linspace(50, 150, 100)
prices = [black_scholes_price(S,K,T,r,sigma, option_type) for K in K_values]
```

```python
plt.figure(figsize =(10,5))
plt.plot(K_values,prices, label = f'{option_type.title()}:Option Price')
plt.xlabel = 'Strike Price (K)'
plt.ylabel = 'Option Price'
plt.title = 'Option Price vs Strike Price'
plt.grid(True)
plt.legend()
plt.show()
```

```python
vols = np.linspace(0.01, 1.0, 100)
prices_vol = [black_scholes_price(S, 100, T, r, vol, option_type) for vol in vols]

plt.figure(figsize=(10, 5))
plt.plot(vols, prices_vol)
plt.xlabel = 'Volatility'
plt.ylabel= 'Option Price'
plt.title ='Option Price vs Volatility'
plt.grid(True)
plt.show()
```

```python
prices = [black_scholes_price(S,K,T,r,sigma, option_type= 'put') for K in K_values]

plt.figure(figsize =(10,5))
plt.plot(K_values,prices, label = f'{option_type.title()}:Option Price')
plt.xlabel = 'Strike Price (K)'
plt.ylabel = 'Option Price'
plt.title ='Option Price vs Strike Price'
plt.grid(True)
plt.legend()
plt.show()
```
